```python
import networkx as nx
import matplotlib.pyplot as plt
import numpy as np
```

```python
N = 1000
x = np.linspace(0,4,50)
y = np.array([])
for c in x:
    G = nx.binomial_graph(N, c/(N-1))
    S = len(max(nx.connected_components(G), key=len))
    y = np.append(y, S)
plt.plot(x, y)
```

```python
N = 100
b = np.array([])
a = np.linspace(0,1,50)
for p in a:
    G = nx.binomial_graph(N, p)
    b = np.append(b, nx.average_clustering(G))
plt.plot(a, b)
```

```python
x = np.logspace(1, 6, dtype='int')
self = np.array([])
for N in x:
    flag = False
    while flag==False:
        seq = np.random.zipf(4, N)
        flag = (np.sum(seq) % 2 == 0)
    G = nx.configuration_model(seq)
    self = np.append(self, nx.number_of_selfloops(G)/G.number_of_edges())
plt.plot(np.log(x), self)
```
